```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

def draw_plots_local(classification_report_list):
    # Define the font properties
    font_path = './Arial.ttf'  
    font_prop = fm.FontProperties(fname=font_path)


    name = {'f1-score': [],
            'precision': [],
            'recall': []}

    size = {'f1-score': [],
            'precision': [],
            'recall': []}

    quantity = {'f1-score': [],
            'precision': [],
            'recall': []}

    unit = {'f1-score': [],
            'precision': [],
            'recall': []}

    df = {'f1-score': [],
            'precision': [],
            'recall': []}

    state = {'f1-score': [],
            'precision': [],
            'recall': []}

    temp = {'f1-score': [],
            'precision': [],
            'recall': []}


    for report in classification_report_list:
        for tag, d in report.items():
            if tag == 'NAME':
                for k, v in d.items():
                    if k != 'support':
                        name[k].append(v)
            
            elif tag == 'QUANTITY':
                for k, v in d.items():
                    if k != 'support':
                        quantity[k].append(v)

            elif tag == 'UNIT':
                for k, v in d.items():
                    if k != 'support':
                        unit[k].append(v)

            elif tag == 'STATE':
                for k, v in d.items():
                    if k != 'support':
                        state[k].append(v)

            elif tag == 'DF':
                for k, v in d.items():
                    if k != 'support':
                        df[k].append(v)

            elif tag == 'SIZE':
                for k, v in d.items():
                    if k != 'support':
                        size[k].append(v)

            elif tag == 'TEMP':
                for k, v in d.items():
                    if k != 'support':
                        temp[k].append(v)

    
    print(temp)

    # classwise f1 score
    f = plt.figure(figsize = (10,6))
    plt.xlabel("No. of Epochs", fontsize=12, fontproperties=font_prop)
    plt.ylabel("F1 Score", fontsize=12, fontproperties=font_prop)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    plt.plot(range(len(name['f1-score'])), name['f1-score'], "o-",  label = 'name')
    plt.plot(range(len(state['f1-score'])), state['f1-score'], "o-",  label = 'state')
    plt.plot(range(len(quantity['f1-score'])), quantity['f1-score'], "o-",  label = 'quantity')
    plt.plot(range(len(unit['f1-score'])), unit['f1-score'], "o-",  label = 'unit')
    plt.plot(range(len(df['f1-score'])), df['f1-score'], "o-",  label = 'df')
    plt.plot(range(len(size['f1-score'])), size['f1-score'], "o-",  label = 'size')
    plt.plot(range(len(temp['f1-score'])), temp['f1-score'], "o-",  label = 'temp')
    plt.legend()
    plt.title("Classwise f1 score per Epochs")
    plt.show()

    f.savefig("./Flair_D2/Plot{}.svg".format('classwise_f1'), format='svg', dpi=300)
    f.savefig("./Flair_D2/Plot{}.jpeg".format('classwise_f1'), format='jpeg', dpi=300)
    plt.close()


    # classwise Precision score
    f = plt.figure(figsize = (10,6))
    plt.xlabel("No. of Epochs", fontsize=12, fontproperties=font_prop)
    plt.ylabel("Precision", fontsize=12, fontproperties=font_prop)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    plt.plot(range(len(name['precision'])), name['precision'], "o-",  label = 'name')
    plt.plot(range(len(state['precision'])), state['precision'], "o-",  label = 'state')
    plt.plot(range(len(quantity['precision'])), quantity['precision'], "o-",  label = 'quantity')
    plt.plot(range(len(unit['precision'])), unit['precision'], "o-",  label = 'unit')
    plt.plot(range(len(df['precision'])), df['precision'], "o-",  label = 'df')
    plt.plot(range(len(size['precision'])), size['precision'], "o-",  label = 'size')
    plt.plot(range(len(temp['precision'])), temp['precision'], "o-",  label = 'temp')
    plt.legend()
    plt.title("Classwise Precision score per Epochs")
    plt.show()

    f.savefig("./Flair_D2/Plot{}.svg".format('classwise_precision'), format='svg', dpi=300)
    f.savefig("./Flair_D2/Plot{}.jpeg".format('classwise_precision'), format='jpeg', dpi=300)

    plt.close()
    

    # classwise Recall score
    f = plt.figure(figsize = (10,6))
    plt.xlabel("No. of Epochs", fontsize=12, fontproperties=font_prop)
    plt.ylabel("Recall", fontsize=12, fontproperties=font_prop)
    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)
    
    plt.plot(range(len(name['recall'])), name['recall'], "o-",  label = 'name')
    plt.plot(range(len(state['recall'])), state['recall'], "o-",  label = 'state')
    plt.plot(range(len(quantity['recall'])), quantity['recall'], "o-",  label = 'quantity')
    plt.plot(range(len(unit['recall'])), unit['recall'], "o-",  label = 'unit')
    plt.plot(range(len(df['recall'])), df['recall'], "o-",  label = 'df')
    plt.plot(range(len(size['recall'])), size['recall'], "o-",  label = 'size')
    plt.plot(range(len(temp['recall'])), temp['recall'], "o-",  label = 'temp')
    plt.legend()
    plt.title("Classwise Recall score per Epochs")
    plt.show()

    f.savefig("./Flair_D2/Plot{}.svg".format('classwise_recall'), format='svg', dpi=300)
    f.savefig("./Flair_D2/Plot{}.jpeg".format('classwise_recall'), format='jpeg', dpi=300)
```

```python
draw_plots_local(classification_report_list)
```

```python
def draw_plots_global(y1, y2, x_name, y_name, title, legend_1,legend_2, x):


    # Define the font properties
    font_path = './Arial.ttf'  
    font_prop = fm.FontProperties(fname=font_path)

   

    f = plt.figure(figsize = (10,6))
    
    # naming the x axis
    plt.xlabel(x_name, fontsize=12, fontproperties=font_prop)
    # naming the y axis
    plt.ylabel(y_name,fontsize=12, fontproperties=font_prop)

    plt.xticks(fontsize=10)
    plt.yticks(fontsize=10)

    #create plot
    plt.plot(range(len(y1)), y1, "o-",  label = legend_1)
    plt.plot(range(len(y2)), y2, "o-",  label = legend_2)
    plt.legend()
    plt.title(title)
    plt.show()

    f.savefig("./Flair_D2/Plot{}.svg".format(x), format='svg', dpi=300)
    f.savefig("./Flair_D2/Plot{}.tiff".format(x), format='tiff', dpi=300)
    f.savefig("./Flair_D2/Plot{}.jpeg".format(x), format='jpeg', dpi=300)
    f.savefig("./Flair_D2/Plot{}.pdf".format(x), format='pdf', dpi=300)
```
